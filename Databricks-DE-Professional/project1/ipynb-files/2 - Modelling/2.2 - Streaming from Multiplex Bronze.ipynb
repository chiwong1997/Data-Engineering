We want to extract all the columns where the topic = orders - lets take a look at this using SQL 

```python
%sql
SELECT v.*
FROM (
  SELECT from_json(cast(value AS STRING), "order_id STRING, order_timestamp TIMESTAMP, customer_id STRING, quantity BIGINT, total BIGINT, books ARRAY<STRUCT<book_id STRING, quantity BIGINT, subtotal BIGINT>>") AS v
  FROM bronze
  WHERE topic = "orders")
```

We want to convert this to a streaming read i.e. a streaming query. 

```python
(spark.readStream
      .table("bronze")
      .createOrReplaceTempView("bronze_tmp"))
```

The function above converts the static table into a streaming temporary view. This lets us write streaming queries with SparkSQL. The temporary view is called bronze_tmp. We can now run the same query as in cell 8, but on the temporary view bronze_tmp instead. This will become a streaming query and will continue running until interrupted. 

```python
%sql
SELECT v.*
FROM (
  SELECT from_json(cast(value AS STRING), "order_id STRING, order_timestamp Timestamp, customer_id STRING, quantity BIGINT, total BIGINT, books ARRAY<STRUCT<book_id STRING, quantity BIGINT, subtotal BIGINT>>") v
  FROM bronze_tmp
  WHERE topic = "orders")
```

We will define the SQL logic into a temporary view to pass it back to Python. The temporary view below is used to capture the query we want to apply to our bronze table to get our silver table. 

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW orders_silver_tmp AS
  SELECT v.*
  FROM (
    SELECT from_json(cast(value AS STRING), "order_id STRING, order_timestamp Timestamp, customer_id STRING, quantity BIGINT, total BIGINT, books ARRAY<STRUCT<book_id STRING, quantity BIGINT, subtotal BIGINT>>") v
    FROM bronze_tmp
    WHERE topic = "orders")
```

Now that we have saved our query into a temporary view, we can create our silver table using Python. 

```python
query = (spark.table("orders_silver_tmp")
               .writeStream
               .option("checkpointLocation", "dbfs:/mnt/demo_pro/checkpoints/orders_silver")
               .trigger(availableNow=True)
               .table("orders_silver"))

query.awaitTermination()
```

```python
from pyspark.sql import functions as F

json_schema = "order_id STRING, order_timestamp Timestamp, customer_id STRING, quantity BIGINT, total BIGINT, books ARRAY<STRUCT<book_id STRING, quantity BIGINT, subtotal BIGINT>>"

query = (spark.readStream.table("bronze")
        .filter("topic = 'orders'")
        .select(F.from_json(F.col("value").cast("string"), json_schema).alias("v"))
        .select("v.*")
     .writeStream
        .option("checkpointLocation", "dbfs:/mnt/demo_pro/checkpoints/orders_silver")
        .trigger(availableNow=True)
        .table("orders_silver"))

query.awaitTermination()
```
